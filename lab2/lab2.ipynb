```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sample_image(image,factor):
    height ,width = image.shape[:2]
    sample_image = cv2.resize(image,(width//factor,height//factor),interpolation=cv2.INTER_AREA)
    return sample_image
def quantize_image(image,levels):
   quantized_image = np.floor(image / (256 // levels)) * (256 //levels)
   quantized_image = quantized_image.astype(np.uint8)
   return quantized_image
def plot_images(original, sampled, quantized):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.subplot(1, 3, 2)
    plt.imshow(sampled, cmap='gray')
    plt.title('Sampled Image')
    plt.axis('off')
    plt.subplot(1, 3, 3)
    plt.imshow(quantized, cmap='gray')
    plt.title('Quantized Image')
    plt.axis('off')
    plt.show() 
```

```python
image_path = 'the sun.jpg'
sampling_factor =14
quantize_levels =9

#task1
original_image=cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if original_image is None:
    print(f"error:unable to load image at {image_path}")

sampled_image=sample_image(original_image, sampling_factor)
quantized_image=quantize_image(original_image,quantize_levels)

plot_images(original_image,sampled_image,quantized_image)


from PIL import Image
img1=Image.open("forest.jpg")
img2=Image.open("space.jpg")

resize=(400,400)
img1=img1.resize(resize,Image.Resampling.LANCZOS)
img2=img2.resize(resize,Image.Resampling.LANCZOS)

im1array=np.array(img1)
im2array=np.array(img2)

addition=im1array+im2array
resultImage=Image.fromarray(addition)
resultImage.show()

img3=Image.open("the moon.jpg")
img4=Image.open("mario.jpg")

img3=img3.resize(resize,Image.Resampling.LANCZOS)
img4=img4.resize(resize,Image.Resampling.LANCZOS)

im3array=np.array(img3)
im4array=np.array(img4)

union=im3array|im4array
resultImage2=Image.fromarray(union)
resultImage2.show()
```

```python
def plt_image(image,title="image",imgsize=(4,4)):
    plt.figure(figsize=imgsize)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```
